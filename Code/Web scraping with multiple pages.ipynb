```python
import requests
from bs4 import BeautifulSoup
```

```python
all_words = []


base_url = "https://www.dictionary.com/list/a/"


for page_number in range(1, 43):

    current_url = f"{base_url}{page_number}"

    req = requests.get(current_url)
    

    if req.status_code == 200:

        soup = BeautifulSoup(req.text, 'html.parser')
        

        word_elements = soup.find_all('a', href=True)
        
        words = [element.text for element in word_elements if element.text.startswith("a") and " " not in element.text]
        all_words.extend(words)
    else:
        print(f"Failed to retrieve page {page_number}")
        continue
        
for word in all_words:
    print(word)
```
