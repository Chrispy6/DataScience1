```python
import requests
from bs4 import BeautifulSoup
import re
url = "https://en.wikipedia.org/wiki/Pen_spinning"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
num_hyperlinks = 0
for tag in soup.find_all():
        if tag.name == 'a':
            num_hyperlinks += 1
num_hyperlinks
```

```python
def count_hyperlinks(soup):
    if soup.name == 'a':
        return 1
    count = 0
    for child in soup.children:
        if child.name:
            count += count_hyperlinks(child)
    return count
count_hyperlinks(soup)
```
