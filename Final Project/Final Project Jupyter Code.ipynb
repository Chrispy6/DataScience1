```python
import pandas
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv("netflix_data.csv")
```

```python
data = df[df['release_year'] >= 2013]
```

```python
data.info()
```

```python
sns.countplot(x='type', data=data)
```

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='release_year', hue='type', data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Distribution of Netflix Movies and Shows from 2013-2021')
plt.show()
```

```python
#ratio increase of tv shows from 2013 vs 2021
df_selected_years = data[(data['release_year'] == 2013) | (data['release_year'] == 2021)]

# Step 2: Calculate the count of TV shows for each year
tv_shows_by_year = df_selected_years[df_selected_years['type'] == 'TV Show'].groupby('release_year').size()

# Step 3: Compute the percentage increase
percentage_increase = ((tv_shows_by_year.loc[2021] - tv_shows_by_year.loc[2013]) / tv_shows_by_year.loc[2013]) * 100

print("Percentage Increase of TV Shows from 2013 to 2021:")
print(percentage_increase)
```

```python
df_selected_years = data[(data['release_year'] == 2013) | (data['release_year'] == 2021)]

# Step 2: Calculate the count of movies for each year
movies_by_year = df_selected_years[df_selected_years['type'] == 'Movie'].groupby('release_year').size()

# Step 3: Compute the percentage increase
percentage_increase_movies = ((movies_by_year.loc[2021] - movies_by_year.loc[2013]) / movies_by_year.loc[2013]) * 100

print("Percentage Increase of Movies from 2013 to 2021:")
print(percentage_increase_movies)
```

```python
movies_df = data[data['type'] == 'Movie']

# Step 2: Calculate the average number of movies
average_movies = movies_df.shape[0] / len(movies_df['release_year'].unique())

print("Average number of movies:", average_movies)
```

```python
tv_shows_df = data[data['type'] == 'TV Show']

# Step 2: Calculate the average number of TV shows
average_tv_shows = tv_shows_df.shape[0] / len(tv_shows_df['release_year'].unique())

print("Average number of TV shows:", average_tv_shows)
```
